# Pandas

## Loading Data

```python
#importar o pandas
import pandas as pd
```

```python
#Usar as pd é apenas um apelido (alias) que facilita o uso das funções da biblioteca.
df = pd.read_csv("dados.csv")
#A função read_csv é uma das mais usadas e permite carregar um arquivo CSV diretamente para um DataFrame.
```

## Data Frames
Um DataFrame é a estrutura de dados principal do Pandas. Ele funciona como uma tabela, composta por:

- Linhas
- Colunas

```python
#Criar dataframe a partir de um CSV
csv_file = "dados.csv"
df = pd.read_csv(csv_file)

#Criar dataframe a partir de um dicionário
dados = {
    "nome": ["Ana", "Bruno", "Carlos"],
    "idade": [20, 25, 30]
}

df = pd.DataFrame(dados)
```

Acessando linhas e colunas

- df["coluna"] → acessa uma coluna

- df.loc[] → acessa por label (nome do índice)

- df.iloc[] → acessa por posição (número da linha)

```python
df["nome"]           # coluna 'nome'
df.loc[0]            # linha com índice 0
df.iloc[1, 0]         # valor na linha 1, coluna 0
```

## Operações com DataFrames

```python
#Selecionar colunas específicas
df_nomes = df[["nome"]]

#Selecionar linhas específicas
primeiras_linhas = df.iloc[0:2]   # linhas 0 e 1

#Slicing (Fatiamento)
df_subset = df.iloc[1:4, 0:2]
```

### Identificação de Elementos Únicos

```python
# unique() retorna todos os valores únicos de uma coluna
anos_unicos = df["Released"].unique()
print(anos_unicos)
```

### Filtragem de Dados

```python
#criar um novo DataFrame com base em uma condição lógica

#selecionar apenas músicas/álbuns lançados após 1979:
df_pos_1979 = df[df["Released"] > 1979]

#df["Released"] > 1979 gera uma série booleana, como:
[True, False, True, True, ...]
#Essa série é usada pelo Pandas para filtrar apenas as linhas onde a condição é verdadeira.
#permite filtrar dados em uma única linha

#Filtrar colunas específicas ao mesmo tempo:
df_filtrado = df[df["Released"] > 1979][["Artist", "Album", "Released"]]
```

### Salvar Resultados
Formatos Suportados:

- CSV → df.to_csv("arquivo.csv")
- Excel → df.to_excel("arquivo.xlsx")
- JSON → df.to_json("arquivo.json")
- HTML → df.to_html("arquivo.html")
- Parquet → df.to_parquet("arquivo.parquet")
- Pickle → df.to_pickle("arquivo.pkl")

```python
#Salvar em CSV
df_pos_1979.to_csv("musicas_pos_1979.csv", index=False)
```
