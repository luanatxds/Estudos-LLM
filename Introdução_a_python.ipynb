## Listas e Tuplas

```python
#Tuplas
tupla = (10, 20, 30)
tupla[1] = 99  # ❌ erro, não pode modificar
```

```python
#listas
lista = [10, 20, 30]
lista.append(40)      # adiciona
lista[1] = 99         # modifica
lista.remove(10)      # remove

print(lista)
```

```python
#a indexação é igual
lista = [10, 20, 30, 40]
tupla = (10, 20, 30, 40)

print(lista[0])  # 10
print(tupla[2])  # 30
print(lista[-1])  # 40 (último elemento)
print(tupla[-2])  # 30

```

```python
#slice
lista = [1, 2, 3, 4, 5]
print(lista[1:4])  # [2, 3, 4]

tupla = (1, 2, 3, 4, 5)
print(tupla[:3])   # (1, 2, 3)

#formato = sequencia[início:fim:passo]

print(lista[::2])  # [1, 3, 5] (pulando de 2 em 2)
print(tupla[::-1]) # (5, 4, 3, 2, 1) (invertida)
```

```python
#tamanho
print(len(lista))
print(len(tupla))
#verificar se existe dentro

lista = ['uva', 2, 3]
tupla = (1, 2, 3)
print("uva" in lista)
print("uva" in tupla)

#concatenar
nova_lista = lista + [10, 20]
nova_tupla = tupla + (10, 20)

print(nova_lista)
print(nova_tupla)
```

```python
#Conversão entre lista e tupla
#Lista → Tupla
lista = [1, 2, 3]
tupla = tuple(lista)
print(tupla)  # (1, 2, 3)

#Tupla → Lista
tupla = (1, 2, 3)
lista = list(tupla)
print(lista)  # [1, 2, 3]
```

```python
#Iteração (loop que printa um item de cada vez)
for item in lista:
    print(item)

for item in tupla:
    print(item)
```

```python
#Desempacotamento
ponto = (3, 7)
x, y = ponto

print(x)  # 3
print(y)  # 7
```

```python
#operações em listas
lista = [1, 2, 3]
lista.append("novo")      #inserir no final
print(lista)

lista.insert(1, "item")   #inserir em posição específica
print(lista)

lista.remove("item")      #remove pelo valor
print(lista)

lista.pop(3)              #remove pelo índice
print(lista)

lista.reverse()           #inverte a ordem
print(lista)

lista.sort()              #ordena
print(lista)

lista.clear()             #remove todos os elementos
print(lista)

lista.extend([4, 5, 6])   #adiciona vários elementos
print(lista)

del(lista[0])             #remove pelo índice
print(lista)

help(lista)               #mostra a documentação
```

## Dicionário

## Conjuntos / Sets
- **Não permite duplicatas**
- **Não é indexado** (não dá para acessar por `[0]`)
- **A ordem não é garantida**
- **Mutável** (pode adicionar e remover itens)
- Aceita apenas **elementos imutáveis** (ex: int, string, tupla)

```python
#lista sem duplicatas
meu_set = set([1, 2, 2, 3])
print(meu_set)  # {1, 2, 3}

meu_set.add(4)
print(meu_set)  # {1, 2, 3, 4}

meu_set.remove(2)
print(meu_set)  # {1, 3, 4}

if 3 in meu_set:
    print("existe")

print(len(meu_set))
```

```python
#União
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)        # {1, 2, 3, 4, 5}
print(a.union(b))   # {1, 2, 3, 4, 5}

#interseção
print(a & b)           # {3}
print(a.intersection(b))

#diferença
print(a - b)           # {1, 2}
print(a.difference(b))

#diferença simétrica
print(a ^ b)           # {1, 2, 4, 5}
print(a.symmetric_difference(b))
```

```python
#Exemplo 1
lista = [1, 2, 2, 3, 3, 3, 4]
unicos = list(set(lista))
print(unicos)

#Exemplo 2
compras1 = {"arroz", "feijão", "pão"}
compras2 = {"pão", "café", "queijo"}

print(compras1 & compras2)  # {'pão'}

#exemplo 3
grupo_a = {"python", "sql", "docker"}
grupo_b = {"python", "react", "figma"}

print(grupo_a ^ grupo_b)
# {'sql', 'docker', 'react', 'figma'}
```

## Tratamento de exceções

evitar que o programa quebre

```python
try:
    numero = int("abc")  # erro
except:
    print("Erro ao converter para inteiro!")
```

```python
#erros específicos

#ZeroDivisionError
#ValueError
#TypeError
#FileNotFoundError
#IndexError
#KeyError
#NameError

try:
    resultado = 10 / 0
except ZeroDivisionError:
    print("Não é possível dividir por zero!")
```

```python
#Múltiplos except
try:
    x = int("abc")
except ValueError:
    print("Erro de valor!")
except TypeError:
    print("Erro de tipo!")
```

```python
#Capturando exceção e lendo a mensagem de erro
try:
    int("abc")
except ValueError as e:
    print("Ocorreu um erro:", e)
```

```python
#bloco finally
try:
    arquivo = open("dados.txt")
    print(arquivo.read())
except FileNotFoundError:
    print("Arquivo não encontrado.")
finally:
    print("Encerrando operação.")
```

```python
#bloco else
try:
    n = int("10")
except ValueError:
    print("Erro ao converter.")
else:
    print("Conversão bem-sucedida!")
```

```python
#Lançando (criando) suas próprias exceções
#gerar um erro manualmente com raise.
def dividir(a, b):
    if b == 0:
        raise ValueError("b não pode ser zero.")
    return a / b
```

## Objetos e Classes

Uma classe é como um molde ou projeto.
Ela define:

atributos → dados

métodos → funções que usam esses dados

Exemplo:
Classe "Carro" → define a estrutura geral de um carro (cor, modelo, velocidade).

Um **objeto** é uma instância da classe, ou seja, uma versão real construída a partir do molde.

```python
class Carro:
    def __init__(self, marca, cor):
        self.marca = marca      # atributo
        self.cor = cor          # atributo

    def buzinar(self):          # método
        print("Beep beep!")
```

### **O método __init__**

É o *construtor*, executado sempre que um novo objeto é criado.
self representa o próprio objeto.

```python
carro1 = Carro("Toyota", "Preto")
carro2 = Carro("Honda", "Vermelho")

print(carro1.marca)   # Toyota
print(carro2.cor)     # Vermelho

carro1.buzinar()      # Beep beep!
```

```python
#Atributos de instância - Pertencem ao objeto
class Carro:
  def __init__(self, marca, cor):
    self.marca
    self.cor

#Atributos de classe - Pertencem à classe inteira - iguais para todos os objetos
class Carro:
    rodas = 4   # atributo de classe
```

```python
#Métodos especiais
# _init_ - construtor
# __str__ - representação em string
# __len__ - usado por len(obj)
# __eq__ - comparação ==

class Pessoa:
    def __init__(self, nome):
        self.nome = nome

    def __str__(self):
        return f"Pessoa: {self.nome}"
```
