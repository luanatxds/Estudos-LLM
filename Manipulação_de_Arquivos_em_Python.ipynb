## Manipulação de Arquivos em Python

**função open()**
```
open(caminho, modo)
```

```
# Isto está formatado como código
```



````python
#Modos comuns
# 'r'	leitura
# 'w'	escrita (sobrescreve)
# 'a'	acrescentar conteúdo ao final
# 'r+'	leitura + escrita
# 'b'	modo binário

# Combinar modos
# 'rb' → ler arquivo binário
# 'w+' → ler e escrever (sobrescreve)

# abrir e fechar arquivo manualmente
arquivo = open("dados.txt", "r")
conteudo = arquivo.read()
arquivo.close()

# abrir e fechar automaticamente
with open("dados.txt", "r") as arquivo:
    conteudo = arquivo.read()
````

````python
#ler arquivos

# read() lê o arquivo inteiro como string
with open("dados.txt", "r") as arquivo:
    conteudo = arquivo.read()
    print(conteudo)

# readline() lê uma linha por vez
with open("dados.txt", "r") as f:
    linha = f.readline()
    print(linha)

linha = f.readline(10)  # lê apenas 10 caracteres

#readlines() retorna uma lista com todas as linhas
with open("dados.txt", "r") as f:
    linhas = f.readlines()
print(linhas)
#['linha 1\n', 'linha 2\n', 'linha 3\n']
````

````python
#Ler linha por linha com loop
with open("dados.txt", "r") as f:
    for linha in f:
        print(linha.strip())
#strip() remove quebras de linha.
````

````python
#Escrever em Arquivos

#write()
with open("saida.txt", "w") as f:
    f.write("Olá, mundo!")

#escrever várias linhas
linhas = ["linha 1\n", "linha 2\n"]

with open("saida.txt", "w") as f:
    f.writelines(linhas)
````

````python
#escrever listas com loop
linhas = ["Apple", "Banana", "Cherry"]

with open("frutas.txt", "w") as f:
    for item in linhas:
        f.write(item + "\n")

#forma alternativa writelines()
linhas = ["Linha 1\n", "Linha 2\n", "Linha 3\n"]

with open("saida.txt", "w") as f:
    f.writelines(linhas)
````
